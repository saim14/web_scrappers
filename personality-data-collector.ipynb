## Details report data download

```python
import requests
from bs4 import BeautifulSoup
ID = '1003'
name = 'Ishraf bin sharif'
test_id = "6065864ad1e9fd000859bc54"
page = requests.get(f"https://bigfive-test.com/result/{test_id}")
soup = BeautifulSoup(page.content, 'html.parser')
print(f"{name}, this is your report")
print("ID:",ID)
print("Name:",name)
print("Test ID",test_id)
print("=================================")
summary_title = ['Neroticism','Extraversion','Openness To Experience','Agreeableness','Conscientiousness']
columns = ["ID", "name", "test_id"]
data = [ID,name,test_id]
for n0 in range(4,9):
    total = 0
    for n1 in range(4,10):
        select_full = soup.select(f'div.v-application--wrap > main > div > div > div > div > div:nth-child({n0}) > div > div:nth-child({n1})')
        title = select_full[0].find('a').get_text().strip()
        columns.append(title)
        score = select_full[0].find('p').get_text().strip().split(" ")[1]
        data.append(int(score))
        total = total + int(score)
        print(f"{title}: {score}")
    print('---------------------------------')
    print(f"\033[1m{summary_title[n0-4]}: {total}\033[0m")
    columns.append(summary_title[n0-4])
    data.append(total)
    print("=================================")
print(columns)
print(data)


import csv

with open('personality.csv', mode='a') as personality_file:
    personality_writer = csv.writer(personality_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    #personality_writer.writerow(columns)
    personality_writer.writerow(data)
```
